## Load CIFAR-10 Data

```python
def load_cifar():
    
    trn_data, trn_labels, tst_data, tst_labels = [], [], [], []
    def unpickle(file):
        with open(file, 'rb') as fo:
            data = pickle.load(fo, encoding='latin1')
        return data
    
    for i in trange(1):
        batchName = './data/data_batch_{0}'.format(i + 1)
        unpickled = unpickle(batchName)
        trn_data.extend(unpickled['data'])
        trn_labels.extend(unpickled['labels'])
    unpickled = unpickle('./data/test_batch')
    tst_data.extend(unpickled['data'])
    tst_labels.extend(unpickled['labels'])
    return trn_data, trn_labels, tst_data, tst_labels
```

## Image preprocessing

```python
def image_prep(image):
    ''' pre-processes the given image
        performs mean normalization and other such operations'''
    image = (image - np.mean(image))/np.std(image)
    processed_image = 2*(image-np.min(image))/(np.max(image)-np.min(image)) - 1
    return processed_image
```

## Dimensionality reduction

```python
def reduce_dim(X_train, X_val, X_test, y_train, y_val, y_test, **kwargs):
    ''' performs dimensionality reduction'''
    if kwargs['method'] == 'pca':
        pca = PCA(n_components=64)
        pca.fit(X_train)
        X_train = pca.transform(X_train)
        X_val = pca.transform(X_val)
        X_test = pca.transform(X_test)
    elif kwargs['method'] == 'lda':
        lda = LDA(n_components=9)
        lda.fit(X_train, y_train)
        X_train = lda.transform(X_train)
        X_val = lda.transform(X_val)
        X_test = lda.transform(X_test)
    elif kwargs['method'] == 'nonLinear':
        kpca = KernelPCA(n_components=32)
        kpca.fit(X_train)
        X_train = kpca.transform(X_train)
        X_val = kpca.transform(X_val)
        X_test = kpca.transform(X_test)
    
    return X_train, X_val, X_test
```

## Classification

```python
def classify(X, y, **kwargs):
    ''' trains a classifier by taking input features
        and their respective targets and returns the trained model'''
    if kwargs['method'] == 'linearSVM':
        clf = SVC(kernel='linear',C=0.1,gamma=(1/(len(X)*np.std(X))))
        clf.fit(X,y)
    elif kwargs['method'] == 'kernelSVM':
        clf = SVC(kernel='rbf', gamma=(1/(len(X)*np.std(X))))
        clf = clf.fit(X, y)
    elif kwargs['method'] == 'logisticRegression':
        clf = LogisticRegression(C = 2.7825594022071245, penalty='l2')
        clf = clf.fit(X,y)
    elif kwargs['method'] == 'MLP':
        clf = MLPClassifier()
        clf = clf.fit(X, y)
    return clf
```

## Evaluation 

```python
def evaluate(target, predicted):
    f1 = f1_score(target, predicted, average='micro')
    acc = accuracy_score(target, predicted)
    return f1, acc
```

```python
def test(clf, X, y, **kwargs):
    '''takes test data and trained classifier model,
    performs classification and prints accuracy and f1-score'''
    f1, acc = evaluate(clf.predict(X), y)
    print('F1 score: {}\t Accuracy: {}'.format(f1, acc))
```

```python
def main():
    trn_data, trn_labels, tst_data, tst_labels = load_cifar()
    trn_data, tst_data = list(map(image_prep, trn_data)), list(map(image_prep, tst_data))
    X_train, X_val, y_train, y_val = train_test_split(trn_data, trn_labels,test_size = 0.20)
    y_test = tst_labels
    ''' perform dimesioality reduction/feature extraction and classify the features into one of 10 classses
        print accuracy and f1-score.
        '''
    X_train, X_val, X_test = reduce_dim(X_train, X_val, tst_data, y_train, y_val, tst_labels, method='nonLinear')
    clf = classify(X_train, y_train, method='logisticRegression')
    test(clf, X_val, y_val)
    test(clf, X_test, y_test)
```
